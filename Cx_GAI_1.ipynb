```python
def poem_on_rain():
    #prompt = 'Write a 4 line poem on Rain'

    response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      messages=[
        {"role": "user", "content": "Write a 4 line poem on Rain"}
        ]
    )
    print(response.choices[0]['message']['content'])

if __name__ == '__main__':
    poem_on_rain()
```
